**Задание 1.** Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
nomer = "А234ВС777"
```

```python
import re
```

```python
def nomer_valid(nomer:str):
    if isinstance(nomer, str):
        if re.match(r'[авекмнорстух]\d{3}[авекмнорстух]{2}\d{2,3}', nomer.lower()):
            return re.findall(r'[авекмнорстух]\d{3}[авекмнорстух]{2}', nomer.lower())[0], re.findall(r'\d{2,3}$', nomer.lower())[0]
    return "Номер не является валидным"        
```

```python
nomer_valid(input("Введите Госномер"))
```

**Задание 3.** Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
def povtor(s):
    return (re.sub( r'(\S+) (?:\1 ?){2,}', r'\1 \1', s))
```

```python
povtor(s)
```

**Задание 4**. Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:   
+7(xxx)-xxx-xx-xx  
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера.
Примеры валидных форматов:
- +7 955 555-55-55;
- 9555555555;
- 8(955)555-55-55;
- +7 955 555 55 55;
- 7(955) 555-55-55;
- +7 955+555+55+55.

```python
def phone_number(d:str):
    if isinstance(d, str):
        if re.match(r"\+?7?8?(\(?|\s?)9\d{2}(\)|\s|\+)\d{3}(\s|-|\+)?\d{2}(\s|-|\+)?\d{2}", d):
            return "Номер телефона является валидным"
    return "Номер не является валидным"    
```

```python
phone_number("+7 955 555 55 55")
```
